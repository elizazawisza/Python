## Zadanie 1 _(5pt)_

Napisz funkcję, który wyświetla trójkąt Pascala o zadanej wysokości
```
                1
              1   1
            1   2   1
          1   3   3   1
        1   4   6   4   1
      .....................
```

````python
def pascal_triangle(n):
    array = []
    for i in range(n):
        array.append([])
        array[i].append(1)
        for j in range(1, i):
            array[i].append(array[i-1][j-1]+array[i-1][j])
        if i!= 0:
            array[i].append(1)        
    return array    

def print_triangle(array):
    n = len(array)
    for i in range(n):
        print((n-i)* "  " + '   '.join(map(str, array[i])))

print_triangle(pascal_triangle(5))
````

## Zadanie 4 _(10pt)_

Napisz funkcję `prime_factors(n)`, która rozkłada $n$ na czynniki pierwsze i jako wynik wraca listę par $[(p_1,\alpha_1), (p_2,\alpha_2), \ldots, (p_k, \alpha_k)]$ taką, że $p_1^{\alpha_1}p_2^{\alpha_2}\ldots p_k^{\alpha_k}$.


````python
from sympy import factorint
def prime_factors(n):
    factors = factorint(n)
    factorsAmount = len(factors)
    it = 1
    print("[", end = "")
    for p, alfa in factors.items():
        if it == factorsAmount:
            print("(" + str(p) + "," + str(alfa) + ")", end ="")
        else:
            print("(" + str(p) + "," + str(alfa) + ")", end =",")
        it+=1   
            
    print("]", end = "")   

    
prime_factors(126)    
````

## Zadanie 5 _(10pt)_

Napisz funkcję `fraczero(n)` liczącą silnie $n!$ dla $n$ od $0$ do $10000$ oraz jako wynik wraca liczbę zer na końcu $n!$.

````python
def fraczero(n):
    for i in range(n):
        print(i , end =" : ")
        print("The amount of 0 at the end of factorial is : ", end ="")
        amountOfZeros = 0
        j=5
        while (i/j >= 1): 
            amountOfZeros += int(i/j) 
            j *= 5
        print(int(amountOfZeros))
       
fraczero(10000)        
    
````

## Zadanie 7 _(10pt)_

Załóżmy, że mamy następującą listę: `L = ['aababacaa', 'cabaabcca', 'aaabbcbacb', 'acababbaab', ...]` wygenerowaną losowo tzn. lista zawiera losowe ciągi znaków o zadanej długości z liter `'a'`, `'b'` i `'c'`. Takie wzorce występują w wielu miejscach np. sekwencje DNA. Użytkownik zawiera swój (wzorzec) ciąg znaków z literami i pustymi miejscami oznaczonymi przez `'*'` np. `"a**a******"`. Wykorzystaj słownik z kluczami zawierającymi indeksy znaków różnych od `'*'` i wartościami liter z wzorca. Napisz program, który znajduje wszystkie sekwencje z listy, które pasuje do podanego wzorca.

````python
import string 
import random 

def generateRandomStrings(size, amount):
    randomList = []
    for i in range(amount):
        generatedString =  ''.join(random.choices("abc", k = size)) 
        randomList.append(generatedString)
    return randomList    

def generateDictionaryFromTemplate(template):
    Dict = {}
    for i in range(len(template)):
        if not template[i] == '*':
            Dict[i] = template[i]
    return Dict

def searchTemplateInList(L, templateDict):
    results = []
    if len(templateDict) > 0:
        for index, value in templateDict.items():
            results.clear()
            for i in range(len(L)):
                if L[i][index] == value: 
                    results.append(L[i])
            templateDict.pop(index) 
            break;
        if len(templateDict) > 0:
            searchTemplateInList(results, templateDict) 
        else: 
            print(results)
    else: 
        print("ooops there is no template")

L = generateRandomStrings(9, 80)
userTemplate = "a***c****"
dictionary = generateDictionaryFromTemplate(userTemplate)
searchTemplateInList(L, dictionary)
````

## Zadanie 8 _(10pt)_

Napisz program, który konwerteruje liczby rzymskie na liczby arabskie. Do zapisu liczb rzymskich używa
się 7 liter: `I=1`, `V=5`, `X=10`, `L=50`, `C=100`, `D=500`, `M=1000`. Opis algorytmu zamiany z i na system rzymski można znaleźć np. [tutaj](http://www.algorytm.org/algorytmy-arytmetyczne/zamiana-z-i-na-system-rzymski.html)

````python
def romanToArabic(number):
    arabic = [1000, 500, 100, 50, 10, 5, 1]
    roman = ['M', 'D', 'C', 'L', 'X', 'V', 'I']
    i = 0
    j = 0
    resultNumber = 0
    length = 0
    length = len(number)
    
    while j<length and i<len(arabic):
        if number[j] == roman[i]:
            resultNumber+=arabic[i]
            j+=1
        elif i%2 == 0 and i<len(arabic)-2 and j<length-1 and number[j] == roman[i+2] and number[j+1] == roman[i]:  
            resultNumber+=arabic[i] - arabic[i+2]
            j+=2
            i+=1
        elif i%2 == 1 and i<len(arabic)-1  and j<length-1 and number[j] == roman[i+1] and number[j+1] == roman[i]:
            resultNumber+=arabic[i] - arabic[i+1]
            j+=2
            i+=1
        else: 
            i+=1
        if i == len(arabic):
            print("An error occured") 
    return resultNumber;

romanToArabic('MMMDCCXXIV')
````

## Zadanie 9* _(5pt)_

Napisz program kalkulator, który pobiera wprowadzone wartości oraz funkcje z klawiatury następnie podaje wynik. W przypadku błędnych danych zgłasza błąd.

Przykładowa sesja:
```
    Kalkulator
    [1]: 2+5*10
        52
    [2]: sin(0.5)+cos(0.3)
        1.434762027729809
    [3]: 2^100
        1267650600228229401496703205376
```

Przykładowy kod pobierania danych od użytkownika:

```
    x = input('Podaj x = ')
```

````python
from math import *

def calculator():
    print("Simple calculator app")
    while True:
        userInput = input()
        if '^' in userInput:
            userInput = userInput.replace('^', '**')
        try:
            result = eval(userInput,{'sqrt': sqrt, 'sin': sin, 'cos': cos })
            print(result)
        except NameError:
            print("An error occured. Function doesn't exist")
        except SyntaxError as error: 
            print("An error occured. There is a mistake in input. Try again ")

calculator()            
````
